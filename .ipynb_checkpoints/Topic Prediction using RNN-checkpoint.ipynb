```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("content_topic_prediction/Topic_prediction_data_train.txt",sep='\t',header=None)
data.columns = ['text','uid','type','timestamp']
labels = pd.read_csv("content_topic_prediction/Topic_prediction_labels_train.txt",sep='\t',header=None)
labels.columns = ['label']
data['label'] = labels['label']
```

```python
# remove rows that have null text
data = data[(data['text'].notnull())]
data=data.reset_index(drop=True)
```

```python
alltext = [i.split(' ') for i in data["text"]]
textlength = [len(i.split(' ')) for i in data["text"]]
```

```python
#find all unique words
all_list =  []
for i in range(len(alltext)):
    all_list.extend(alltext[i])
unique=list(set(all_list))
print(len(unique))
    
```

```python
#build a word dictionary
w_dic ={}
for i in range(len(unique)):
    w_dic[unique[i]]=i
```

```python
w_dic
```

```python
def convert_to_int(words):
    converted=[]
    for w in words:
        if w in w_dic:
            converted.append(w_dic[w])
    return converted   
```

```python
data_conv =[]
for item in alltext:
    data_conv.append(convert_to_int(item))
```

```python
print(data_conv[81000])
print(alltext[81000])
```

```python
print(w_dic["0url1"])
```

```python
data_conv= np.array(data_conv)
print(data_conv.shape)
```

```python
from keras.preprocessing import sequence
# pad
data_conv_trunc= sequence.pad_sequences(data_conv,padding='post')
```

```python
#truncate to the first 250 words
data_conv_trunc= data_conv_trunc[:,:250]
```

```python
print(data_conv_trunc.shape)
```

```python
from keras.utils import np_utils
labels = np.array(data["label"])
labels = np_utils.to_categorical(labels, num_classes=10)
labels.shape
```

```python
from sklearn.model_selection import train_test_split
x_Train, x_Test, y_Train,  y_Test= train_test_split(data_conv_trunc,labels, test_size=.2, stratify = labels)
```

```python
x_Val, x_Test, y_Val,  y_Test= train_test_split(x_Test,y_Test, test_size=.5)
```

```python
from keras.layers import LSTM
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers.embeddings import Embedding
embedding_vecor_length = 32
model = Sequential()
model.add(Embedding(5500+1,32, input_length=250))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

print(model.summary())
```

```python
from keras.callbacks import ModelCheckpoint
checkpointer = ModelCheckpoint(filepath='model.weights.best.hdf5', verbose=1, 
                               save_best_only=True)
model.fit(x_Train, y_Train, validation_data=(x_Val, y_Val), epochs=5, batch_size=64, callbacks=[checkpointer])
```
